```python
not_cat = df[["totalScanTimeInSeconds", "grandTotal", "scannedLineItemsPerSecond", "valuePerSecond", "lineItemVoidsPerPosition"]]
not_cat
```

```python
# рисую разброс только некатегойных данных
for col in not_cat.columns:
    f, ax = plt.subplots(figsize=(11, 6))
    ax.set_title(col)
    sns.boxenplot(data=df[col])
    sns.despine(left=True, bottom=True)
```

```python
np.abs(not_cat.corr()).style.background_gradient(cmap='coolwarm')
```

```python
np.abs(df.corr()).style.background_gradient(cmap='coolwarm')
```

```python
df = df.loc[df["valuePerSecond"] < np.percentile(df['valuePerSecond'], 95, interpolation = 'midpoint')]
not_cat = df[["totalScanTimeInSeconds", "grandTotal", "scannedLineItemsPerSecond", "valuePerSecond", "lineItemVoidsPerPosition"]]
df
```

```python
np.abs(df.corr()).style.background_gradient(cmap='coolwarm')
```

```python
for col in not_cat.columns:
    f, ax = plt.subplots(figsize=(11, 6))
    ax.set_title(col)
    sns.boxenplot(data=df[col])
    sns.despine(left=True, bottom=True)
```

при удалении выбросов повысилась корреляция, интересно, но связь всё равно не достаточно сильная, чтобы делать выводы

```python
for col in not_cat.columns:
    stats.probplot(df[col], dist=stats.distributions.norm, plot=pylab, sparams=(2))
    pylab.title(col)
    pylab.show()
```
